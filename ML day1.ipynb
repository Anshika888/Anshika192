```python
plt.figure(figsize=(8,6))
plt.plot(def_tips["day"],def_tips["total_bill"],marker="o",linestyle="--",color="blue")
plt.title("Line Plot:Total Bill Across Days")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill($)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(def_tips["day"].unique(),def_tips.groupby("day")["total_bill"].mean(),color="green")
plt.title("Bar Plot:Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel(" Average Total Bill($)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(def_tips["total_bill"],def_tips["tip"],color="orange")
plt.title("scatter Plot:Total Bill Vs Tip")
plt.xlabel("Total Bill (rs)")
plt.ylabel("Tip (rs)")
plt.show()
```
